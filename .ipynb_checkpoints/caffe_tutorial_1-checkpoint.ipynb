### Loading the model

Here is where we load the models and weights by using the caffe.Net(). caffe.TEST of this line means we will be making our classification, as opposed to training it.

##### Set the mode to CPU mode
##### Load model architecture from path stored in model_def
##### Load model pre-trained weights from the file we just download

```python
caffe.set_mode_cpu()

model_def = caffe_root + 'models/bvlc_reference_caffenet/deploy.prototxt'
model_weights = caffe_root + 'models/bvlc_reference_caffenet/bvlc_reference_caffenet.caffemodel'

net = caffe.Net(model_def, model_weights, caffe.TEST)
```

### Preprocess the data into desired format

#### Image Processing Techniques used here
Subtracting mean values - 
Transposing - 
Channel Swapping - caffe requires BGR for their inputs

Caffe Documentation here: 

```python
mu = np.load(caffe_root + 'python/caffe/imagenet/ilsvrc_2012_mean.npy')
mu = mu.mean(1).mean(1)
print 'mean-subtracted values: ', zip('BGR', mu)

transformer = caffe.io.Transformer({'data': net.blobs['data'].data.shape})

transformer.set_transpose('data', (2, 0, 1))
transformer.set_mean('data', mu)
transformer.set_raw_scale('data', 255)
transformer.set_channel_swap('data', (2, 1, 0))
```

### The final format of our data
#### Blob?
A blob is a wrapper over the actual data being processed and passed along by Caffe. Mathematically, it is an N-dimensionary array. Blobs are Caffe's way of storing and communicating data. 

Blob dimensions go in this format:
(# of data, # of channel, width, height)

Parameter blob dimensions vary according to the type and configuration of the layer. 

```python
tuple(net.blobs['data'].shape)
```

```python
net.blobs['data'].reshape(50, 3, 227, 227)
```

### An example of the processed image

```python
image = caffe.io.load_image(caffe_root + 'examples/images/cat.jpg')
transformed_image = transformer.preprocess('data', image)
plt.imshow(image)
```

### Prediction of the cat image above

net.forward() is the essential computation of the net. This is where it computes the output, given the input for inference. Computation is done in each layer to compute the "function". In other words, it will yield the output and loss. 

```python
net.blobs['data'].data[...] = transformed_image
output = net.forward()
output_prob = output['prob'][0]
print 'predicted class is:', output_prob.argmax()
```

### Interpretation of the label - 281 corresponds to tabby cat

```python
labels_file = caffe_root + 'data/ilsvrc12/synset_words.txt'
if not os.path.exists(labels_file):
    !/data5/guest/caffe/data/ilsvrc12/get_ilsvrc_aux.sh
    
labels = np.loadtxt(labels_file, str, delimiter='\t')

print'output label:', labels[output_prob.argmax()]
```

### Top 5 prediction - A common use metrics

```python
top_ind = output_prob.argsort()[::-1][:5]
print 'probability and labels'
zip(output_prob[top_ind], labels[top_ind])
```

### The shape of the output from each layers
1. conv - Convolutional Layer
2. pool - Max-Pooling Layer
3. norm - Local Response Normalization Layer
4. fc - Fully Connected Layer
5. prob - Output of Probabilities for Each of 1000 Classes

```python
for layer_name, blob in net.blobs.iteritems():
    print layer_name + '\t' + str(blob.data.shape)
```

### Network Setting
    (a, b, c, d)
    a - Number of filters in conv, and number of neurons in fc
    b - For conv, it is depth of filters in conv (since it is conv2D, it is equal to depth of input)
    b - For fc, it is simply the input dimension
    c & d - filter width and height

```python
for layer_name, param in net.params.iteritems():
    print layer_name + '\t' + str(param[0].data.shape), str(param[1].data.shape)
```

```python
def vis_square(data):
    """Take an array of shape (n, height, width) or (n, height, width, 3)
       and visualize each (height, width) thing in a grid of size approx. sqrt(n) by sqrt(n)"""
    
    # normalize data for display
    data = (data - data.min()) / (data.max() - data.min())
    
    # force the number of filters to be square
    n = int(np.ceil(np.sqrt(data.shape[0])))
    padding = (((0, n ** 2 - data.shape[0]),
               (0, 1), (0, 1))                 # add some space between filters
               + ((0, 0),) * (data.ndim - 3))  # don't pad the last dimension (if there is one)
    data = np.pad(data, padding, mode='constant', constant_values=1)  # pad with ones (white)
    
    # tile the filters into an image
    data = data.reshape((n, n) + data.shape[1:]).transpose((0, 2, 1, 3) + tuple(range(4, data.ndim + 1)))
    data = data.reshape((n * data.shape[1], n * data.shape[3]) + data.shape[4:])
    
    plt.imshow(data); plt.axis('off')
```

### Visualizing the convolutional layer 1
It learns very simple features because the filter can only look into small area

```python
filters = net.params['conv1'][0].data
vis_square(filters.transpose(0, 2, 3, 1))
```

### A complete process of how to predict an image

```python
# download an image
my_image_url = "http://pic2.cxtuku.com/00/00/24/b77547da0628.jpg"  # paste your URL here
# for example:
# my_image_url = "https://upload.wikimedia.org/wikipedia/commons/b/be/Orang_Utan%2C_Semenggok_Forest_Reserve%2C_Sarawak%2C_Borneo%2C_Malaysia.JPG"
!wget -O image.jpg $my_image_url

# transform it and copy it into the net
image = caffe.io.load_image('image.jpg')
net.blobs['data'].data[...] = transformer.preprocess('data', image)

# perform classification
net.forward()

# obtain the output probabilities
output_prob = net.blobs['prob'].data[0]

# sort top five predictions from softmax output
top_inds = output_prob.argsort()[::-1][:5]

plt.imshow(image)

print 'probabilities and labels:'
zip(output_prob[top_inds], labels[top_inds])
```
